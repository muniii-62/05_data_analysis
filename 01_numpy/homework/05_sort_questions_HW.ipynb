### [실습 2] 2차원 배열 축 기준 정렬


다음 2차원 배열을 행과 열을 기준으로 각각 정렬하세요.


```python
arr2d = np.array([[8, 3, 5], [2, 9, 1]])
```

````python
arr2d = np.array([[8, 3, 5], [2, 9, 1]])
print(np.sort(arr2d, axis=1)) # g행기준
print(np.sort(arr2d, axis=0))# 열기준
````

### [실습 3] 인덱스를 반환하는 정렬


다음 1차원 배열에서 정렬된 결과의 인덱스를 반환하고, 그 인덱스를 사용하여 배열을 정렬하세요.


```python
arr = np.array([15, 10, 20])
```

````python
arr = np.array([15, 10, 20])

idx = np.argsort(arr)
print(idx)        # [1 0 2]
print(arr[idx])   # [10 15 20]
````
