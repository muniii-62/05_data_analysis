### [실습] 데이터를 기반으로 배열 연산 수행


다음과 같은 과일 판매량 데이터가 있습니다. 각 과일의 1분기와 2분기 판매량이 주어졌을 때, NumPy 배열을 사용하여 아래 문제를 해결하세요.

1. 1분기와 2분기의 과일별 총 판매량을 구하세요.
2. 과일별 평균 판매량을 구하세요.

- **데이터**:
    - 1분기 판매량: `[200, 340, 150, 400]` (사과, 배, 오렌지, 바나나)
    - 2분기 판매량: `[220, 300, 180, 420]` (사과, 배, 오렌지, 바나나)

````python
import numpy as np
sale_1 = [200, 340, 150, 400]
sale_2 = [220, 300, 180, 420]
arr1 = np.array(sale_1)
arr2 = np.array(sale_2)
print('1분기 총 판매량',arr1.sum())
print('2분기 총 판매량',arr2.sum())
print('과일병 평균 판매량:',(arr1+arr2)/2)

````

### [실습 1] 기본 산술 연산


다음 두 배열을 사용하여 각각의 요소별로 덧셈, 뺄셈, 곱셈, 나눗셈을 수행하세요.


```python
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])
```

````python
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])
answer1=arr1+arr2
print('answer1(+) :',answer1)
answer2=arr1-arr2
print('answer2(-) :',answer2)
answer3=arr1 * arr2
print('answer3(*) :',answer3)
answer4=arr1 / arr2
print('answer4(/) :',answer4)
````

### [실습 4] 비교 연산 및 집계 함수 사용


1. 배열에서 20보다 큰 값을 찾아 출력하세요.
2. 배열의 합계, 최대값, 최소값, 평균을 계산하세요.

```python
arr = np.array([10, 20, 30, 40, 50])
```

````python
arr = np.array([10, 20, 30, 40, 50])

print('20보다 큰값:',arr[arr>20])
print('합계:',np.sum(arr))
print('최대값:',np.max(arr))
print('최소값:',np.min(arr))
print('평균:',np.mean(arr))
````
