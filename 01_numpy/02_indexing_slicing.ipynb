# Indexing | Slicing
- indexing: 특정 위치의 값 선택임
- slicing: 범위 단위의 값 선택임
- fancy indexing: 여러 위치를 한 번에 선택하는 방식임
- boolean indexing: 조건이 `True`인 값만 선택하는 방식임


```python
# NumPy import
import numpy as np
from sklearn.externals.array_api_compat.cupy import reshape
```

## indexing
- 인덱스 번호로 원소 하나를 선택함
- 0번부터 시작함
- 음수 인덱스 사용 가능함


```python
# 1차원 ndarray 인덱싱
arr = np.arange(1,11) # 1~10 생성
print('arr:', arr)
print('arr[0],arr[1],arr[2] : ', arr[0], arr[1], arr[2])
print('arr[-1],arr[-2],arr[-3] : ', arr[-1], arr[-2], arr[-3])
```

```python
# 2차원 인덱싱
arr2 = arr.reshape(2,5) # 2행 5열 2차원 배열로 변환
print(arr2)

print('arr2[0][2] :', arr2[0][2])
print('arr2[1][3] :', arr2[1][3])
print('arr2[-2][3] :', arr2[-2][3])
```

## fancy indexing
- 인덱스 목록으로 여러 원소를 선택함
- 필요한 위치만 골라낼 때 사용함
- 반복문 없이 여러 값을 한 번에 추출 가능함


```python
arr = np.linspace(5,50,10,dtype=int)
print('arr : ',arr)
print('arr[[1,2,8]] : ',arr[[1,2,8]])

#list 변수를 이용한 인덱싱
indexes = [3,5,7]
print('arr[indexes]: ',arr[indexes])
```

```python
# 2차원 fancy indexing
arr = np.linspace(5,100,20,dtype=int).reshape(4,5)
print(arr)

print('arr[[0]]:\n',arr[0])
print('arr[[0,1]]:\n',arr[[0,1]])
# arr[[0,1],[1,3]] 코드의 뜻은
# 행과 열을 나타내는 list 에서
# 같은 인덱스의 값을 얻어와 2차원의 좌표를 만들어서 해당 위치 값을 반환
# [[0,1],[1,3]] -> [0][1],[1][3]
print('arr[[0,1],[1,3]]:\n',arr[[0,1],[1,3]]) # [10 45]

print('arr[[0,1],1:4]\n : ',arr[[0,1],1:4])

```

```python
print("arr[[1, 3], 3:]:\n", arr[[1, 3], 3:])

print("arr[[1, 3], -2:]\n", arr[[1, 3], -2:])
```

### boolean을 이용한 fancy indexing
- 조건식 결과가 `True`인 원소만 선택함
- 데이터 필터링의 기본 방식임
- Pandas 조건 필터링과 연결되는 개념임


```python
arr = np.array([1,2,3,4,5])
print('arr[[1,3]] : ', arr[[1,3]])
print(arr[[False,True,False,True,False]])
print('arr % 2 : ',arr % 2)
print('arr % 2==0 : ',arr % 2==0)
print('arr[arr % 2==0] :',arr[arr % 2==0])
```
